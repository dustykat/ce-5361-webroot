````python
# cumulative/incremental functions

# get slope from two points in XY space
def slope(f1,f2,x1,x2):
    slope = (f2-f1)/(x2-x1)
    return slope

# cumulative to incremental
def disaggregate(f,x):
    n=len(x) # length of vectors
    dfdx=[0 for i in range(n)]; # dfdx starts as list of zeros
    for i in range(1,n):
        dfdx[i] = slope(f[i-1],f[i],x[i-1],x[i])
    return dfdx

# incremental to cumulative
def aggregate(vector1,vector2):
    n=len(vector1)
    # fill vector2 with zeros
    vector2 = [0 for i in range(n)]
    vector2[0] = vector1[0]+0.0
    for i in range(1,n):
        vector2[i] = vector2[i-1] + vector1[i]
    return vector2
````

````python
# compute incremental values
incremental_rain = []
incremental_rain = disaggregate(depthP,timeP)
incremental_flow = []
incremental_flow = disaggregate(depthQ,timeQ)
# plot the data
xlabel = 'Time (hours)'
ylabel = 'Incremental Depth (inches/increment)'
legend1 = 'Runoff'
legend2 = 'Precipitation'
ptitle = 'Plot of Rainfall-Runoff for STA08055700 Storm June 17, 1966'

Plot2Lines(timeQ,incremental_flow,timeP,incremental_rain,ptitle,xlabel,ylabel,legend1,legend2);
````

- Lets improve things by interpolating onto hourly time increments.  This would be a necessary step if using Least Squares approach to unit hydrograph generation

````python
import numpy as np
from scipy.interpolate import interp1d

def interpolate_xy_pairs(x_values, y_values, new_x_values):
    """
    Interpolate a set of X-Y pairs onto a new list of uniformly spaced X values.
    
    Parameters:
        x_values (list): List of original X values.
        y_values (list): List of original Y values.
        new_x_values (list): List of uniformly spaced new X values for interpolation.
        
    Returns:
        list: List of interpolated Y values corresponding to new_x_values.
    """
    # Perform linear interpolation
    interp_func = interp1d(x_values, y_values, kind='linear')
    
    # Interpolate onto new X values
    interpolated_y_values = interp_func(new_x_values)
    
    return interpolated_y_values

t_model = [i for i in range(int(bigtime+1))]

    
# Interpolate
q_model = interpolate_xy_pairs(timeQ, incremental_flow, t_model)
p_model = interpolate_xy_pairs(timeP, incremental_rain, t_model)

ylabel = 'Incremental Depth (inches/hour)'
Plot2Lines(t_model,q_model,t_model,p_model,ptitle,xlabel,ylabel,legend1,legend2);
````

We will use these model data for the unit hydrograph, a useful kernel function is $u(t-\tau)=$

````python
# kernel (unit hydrograph) function - this is supposed to be a scalar function, don't feed it lists
def kernelfn(k0,k1,k2,k3,minutes):
    import math
    #k0 =  2z_0 # a scale constant 
    #k1 =  2N-1  # a shape parameter, should be integer
    #k2 =  t_bar # a residence time parameter
    #k3 =  N2    # dispersion parameter (should be integer)
    kernelfn = k0*(1.0/math.gamma(0.5*(k1+1)))*((2.0/k2)*(minutes/k2)**k1)*math.exp(-(minutes/k2)**k3)
    return kernelfn
````

````python
# discrete convolution function
def convolve(N1,excitation,kernel):
# N1 is length of kernel list (already populated)
# excitation is list of inputs (aka rainfall)
# kernel is list of unit responses
# response is convolved response
    response = [0.0 for i in range(N1)] # populate response vector with zeros
    for i in range(N1):
        for j in range(i,N1):
            response[j] = excitation[i]*kernel[(j-i)]+response[j]
    return response
````

````python
# need excess rain - heres where loss model goes
# lets be lazy and use proportional observations Q/P from cumulative
proportion = depthQ[-1]/depthP[-1] # recall these are last values in cumulative space
excess_rain = proportion*p_model
# now build the unit hydrograph
N1 = len(excess_rain)
kernel = []
### parameters by trial-and-error
k0 = 10 # a scale constant - has units of inches/minute
k1 = 5 # a shape parameter
k2 = 400 # timing parameter - has units of minutes
k3 = 2 # dispersion parameter
# the actual UH kernel
for i in range(N1):
    kernel.append(kernelfn(k0,k1,k2,k3,t_model[i]))
# the convolution
response = convolve(N1,excess_rain,kernel)
# plotting results
legend1 = 'Observed Hydrograph'
legend2 = 'Model Hydrograph'
Plot2Lines(t_model,q_model,t_model,response,ptitle,xlabel,ylabel,legend1,legend2);
````

Build a CN model here is a script using method described at [Creating_excess_precipitation](https://serc.carleton.edu/hydromodules/steps/creating_excess.html)

````python
#CN loss model
def xsrain(cumP,CN):
    S = (1000/CN)-10.0
    Ia = 0.2*S
    if cumP >= Ia:
        xsrain=S*(cumP-Ia)/(cumP-Ia+S)
    else:
        xsrain=0.0
    return xsrain
````

````python
# Apply CN model
CN = 78
excess_p = [0 for i in range(len(pd_model))]
for i in range(len(pd_model)):
    excess_p[i] = xsrain(pd_model[i],CN)
# Convert into rates for unitgraph  
excess_rain = []
excess_rain = disaggregate(excess_p,t_model)
#excess_rain

incremental_flow = []
incremental_flow = disaggregate(qd_model,t_model)


# lets plot
xlabel = 'Time (hours)'
ylabel = 'Rate (inches/hour)'
legend1 = 'Runoff'
legend2 = 'Precipitation'
ptitle = 'Plot of Rainfall-Runoff for STA08055700 Storm June 17, 1966'

Plot2Lines(t_model,incremental_flow,t_model,excess_rain,ptitle,xlabel,ylabel,legend1,legend2);
````

````python
import numpy as np
from scipy.interpolate import interp1d

def interpolate_xy_pairs(x_values, y_values, new_x_values):
    """
    Interpolate a set of X-Y pairs onto a new list of uniformly spaced X values.
    
    Parameters:
        x_values (list): List of original X values.
        y_values (list): List of original Y values.
        new_x_values (list): List of uniformly spaced new X values for interpolation.
        
    Returns:
        list: List of interpolated Y values corresponding to new_x_values.
    """
    # Perform linear interpolation
    interp_func = interp1d(x_values, y_values, kind='linear')
    
    # Interpolate onto new X values
    interpolated_y_values = interp_func(new_x_values)
    
    return interpolated_y_values
````

````python
# re-map into minutes
tt_model = [i for i in range(int(60*t_model[-1]+1))]
# re-express t_model in minutes for UH analysis
t_min = [0 for i in range(len(t_model))]
for i in range(len(t_model)):
    t_min[i]=t_model[i]*60.0
    
#print(t_min)
#print(tt_model)
    
# Interpolate
q_model = interpolate_xy_pairs(t_min, incremental_flow, tt_model)
p_model = interpolate_xy_pairs(t_min, excess_rain, tt_model)
legend1 = "Runoff (in/min)"
legend2 = "Excess Precip (in/min)"
ptitle = " Plot of Rainfall-Runoff for STA08055700 Storm June 17, 1966 \n Excess is for CN ="+repr(CN)+" \n add more title"
Plot2Lines(tt_model,q_model,tt_model,p_model,ptitle,xlabel,ylabel,legend1,legend2);
````

Now have excess, and runoff in same units and same time scale - lets convolve and fit a kernel.  Again reusing code:

````python
# now build the unit hydrograph
N1 = len(p_model) #note the name change - for final code get consistent
kernel = []
### parameters by trial-and-error
k0 = 6.8 # N a scale constant - has units of inches/minute
k1 = 5.4 # 2N-1 a shape parameter
k2 = 450 # timing parameter - has units of minutes
k3 = 2*(k1+1) #beta dispersion parameter
# the actual UH kernel
for i in range(N1):
    kernel.append(kernelfn(k0,k1,k2,k3,tt_model[i]))
# the convolution
response = convolve(N1,p_model,kernel)
# plotting results
legend1 = 'Observed Hydrograph'
legend2 = 'Model Hydrograph'
Plot2Lines(tt_model,q_model,tt_model,response,ptitle,xlabel,ylabel,legend1,legend2);
````
